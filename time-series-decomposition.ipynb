```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("airline_passengers.csv")
df.head()
```

```python
df.dtypes
```

```python
df["Month"] = pd.to_datetime(df["Month"], format= "%Y-%m")
```

```python
# set month column as index column
df.set_index("Month", inplace=True)
df.head()
```

```python
df.dtypes
```

```python
# check missing values
df.isnull().sum()
```

```python
df.plot()
plt.show()
```

```python
# There are two types of models - additive and multiplicative
# use additive when the tread is linear otherwise use multiplicative for non linear or exponential trend and season 
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
# now we will do the decomposition. Break this plot into trend seasonal and residual
decom = seasonal_decompose(df, model= "multiplicative")
```

```python
plt.figure(figsize=(12, 4))
decom.plot();
```
